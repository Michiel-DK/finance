```python
import chromadb
from chromadb.utils import embedding_functions

chroma_client = chromadb.Client()
client = chromadb.PersistentClient(path="../chroma_db/")
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
```

```python
chroma_client.delete_collection('transcripts_mililm_l6_v2')
collection = chroma_client.get_or_create_collection(name="transcripts_mililm_l6_v2", embedding_function=sentence_transformer_ef)
```

```python
from langchain.text_splitter import CharacterTextSplitter, RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200)
```

```python
from langchain.docstore.document import Document

documents = []
ids = []
metadatas = []

for transcript in response_ls:
    document = transcript['content']
    docs = text_splitter.split_text(document)
    id = str(transcript['year'])+'Q'+str(transcript['quarter'])+transcript['symbol']
    metadata = {'symbol':transcript['symbol'], 'quarter':transcript['quarter'], \
            'year':transcript['year']}
    
    
    for i, doc in enumerate(docs):
        doc = Document(page_content=doc, metadata=metadata)
        documents.append(doc)
        metadatas.append(metadata)
        ids.append(f'{id}_{i}')
```

```python
documents = [text_splitter.split_text(i['content']) for i in response_ls]
ids = [str(i['year'])+'Q'+str(i['quarter'])+i['symbol'] for i in response_ls]
metadatas = [{'symbol':i['symbol'], 'quarter':i['quarter'], \
            'year':i['year']}\
        for i in response_ls]
```

```python
documents[0]
```

```python
from langchain_community.embeddings import SentenceTransformerEmbeddings


embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain.vectorstores import Chroma


db2 = Chroma.from_documents(documents, embedding_function, persist_directory="./chroma_db")
```

```python
docs_and_scores = db2.similarity_search_with_score("covid", k=2, fetch_k=5)
```

```python
docs_and_scores
```

```python
collection.add(
    documents=documents,
    metadatas=metadatas,
    ids=ids)
```

```python
results = collection.query(
    query_texts=["solar energy demand"],
    n_results=10
)
results
```
