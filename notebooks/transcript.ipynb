```python
import requests

response_ls = []

stock_ls = ['ADBE', 'MSFT', 'ENPH', 'GOOG', 'META', 'FSLR', 'ABT', 'SPGI']
year_ls = [2022,2023]

for s in stock_ls:
    response = requests.get(f'https://financialmodelingprep.com/api/v4/batch_earning_call_transcript/{s}?year=2023&apikey=pD5T46reMptGBcW1hP2PEzQnGAlmBIAM').json()
    response_ls.append(response)
```

```python
import itertools

response_ls = list(itertools.chain(*response_ls))
```

```python
len(response_ls)
```

```python
import chromadb
from chromadb.utils import embedding_functions

chroma_client = chromadb.Client()
client = chromadb.PersistentClient(path="../chroma_db/")
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
```

```python
chroma_client.delete_collection('transcripts_mililm_l6_v2')
collection = chroma_client.create_collection(name="transcripts_mililm_l6_v2", embedding_function=sentence_transformer_ef)
```

```python
documents = [i['content'] for i in response_ls]
ids = [str(i['year'])+'Q'+str(i['quarter'])+i['symbol'] for i in response_ls]
metadatas = [{'symbol':i['symbol'], 'quarter':i['quarter'], \
            'year':i['year']}\
        for i in response_ls]
```

```python
ids
```

```python
collection.add(
    documents=documents,
    metadatas=metadatas,
    ids=ids)
```

```python
results = collection.query(
    query_texts=["solar energy demand"],
    n_results=10
)
results
```
